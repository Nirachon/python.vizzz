สร้าง function ชื่อ createDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน และ ค่า True/False
- ถ้า True จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้า False จะสร้าง dict ที่มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น

เช่น

output1 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )

จะได้ output1 เป็น { 'apple' : 1 , 'banana' : 2 }

output2 = createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )

จะได้ output2 เป็น { 1 : 'apple' , 2 : 'banana' }

```python
def  createDict(listtext, listnumber , logicV):  #สร้างตัวแปรเพื่อรับค่าทั้ง3ตัว createDictชื่อฟังก์ชัน
        if logicV == True:        #logicV == True ค่ารับมาเป็นทรู 
                retdict ={ }
                for index in range(len(listtext)) : # วนตามจำนวนสมาชิก
                     retdict[listtext[index]] = listnumber[index] 
        else:
                retdict ={ }
                for index in range(len(listtext)) :
                     retdict[listnumber[index]] = listtext[index]
                    
        return retdict
                 
```

key index  ที่เอาไว้ชี้

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , True )  #เขียนแทนลงไปในที่สร้าง
```

```python
createDict( [ 'apple' , 'banana' ] , [ 1 , 2 ] , False )
```

# 2.2 แต่ละกรดอะมิโน มีอักษรแต่ละตัวเท่าใด

```python
def countChar(seq,character): #row1,ตัอักษรแต่ละตัว
    charcount = 0 
    for char in seq:
        if char == character: 
            charcount = charcount + 1 # เอาตัวที่นับเจอไปบวกกับตัวนับไปแล้ว
    return charcount
```

```python
countChar(DNAseq.iloc[1][0],'A')  #อยากรู้แถว1มีเอกี่ตัวว0คือคอลัมน์ DNAseqชื่อตาราง
```

```python
ANum = 0 #หาความยาว
TNum = 0
GNum = 0
CNum = 0
for row in range (DNAseq.shape[0]):      #rang (len(DNAseq))
    ANum += countChar(DNAseq.iloc[row][0],'A')
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')

print(f'There are  {ANum} As, {TNum} Ts , {GNum} Gs , {CNum} Cs ')
```

# 2.3 หาสตริงติดกัน5ตัว วิธีทำ ต้องวนลูป ซ้ำ5ตัวมีเยอะเท่าไหร่

```python
def populatate5charstring(seq):
    str_list = list()
    for i in range(len(seq)-4) : #หา5 -4
        str_list.append(seq[i:i+5]) #จะเรียงสาย5ตัวตามโจทย์ที่ให้
        
    return str_list
```

```python
populatate5charstring('ATTAAAGGTTTA')
```

```python
# ตัดแบ่ง5ตัวในโจทย์จากฟังก์ชันข้างบน
```

```python
 populatate5charstring(DNAseq.iloc[2][0]) #แถวที่3
```

```python
LIST_ALL_5CHAR = list() #สร้างลิสเปล่าเพื่อเก็บ
for row in range (DNAseq.shape[0]):
    LIST_ALL_5CHAR = LIST_ALL_5CHAR + populatate5charstring(DNAseq.iloc[row][0]) #เอาลิสว่าบวกฟังก์ชันของเรา ตัดจากตารางดออ จากโลคอลัมน์ที่0
```

```python
len(LIST_ALL_5CHAR) #รู้ว่ามีเท่าไหร่
```

```python
unique_5char = (set(LIST_ALL_5CHAR)) # ตัดตัวที่ซ้ำออก
```

```python
len(unique_5char)
```

ถ้าความยาวไม่ลดลง ไม่ต้องทำต่อ

```python
dict_5char = {}
for sub_str in unique_5char: #ให้ซับวนตามจำนวนยูนิค วน1022
    num = 0
    for s in LIST_ALL_5CHAR: #ค่าทั้งหมดที่ยังไม่เอาตัวซ้ำออก วน28191 ถ้า28191
        if sub_str == s: #ถ้าสับสต เท่ากันให้นัมบวกไป1
            num += 1
    dict_5char[sub_str] = num
```

```python
dict_5char['TGGTG']
```

```python
dict_5char #แต่ละตัวซ้ำกี่ตัว
```

```python
maxVal = -100 #ตั้งน้อยๆไว้ก่อน ตัวเลข13 ถ้าหาตัวน้อยให้ตั้งมาก
maxKey = None #ตัวอักษร CTGGG
for k in dict_5char.keys(): #  dict_5char ชื่ออันก่อนหน้า
    if dict_5char[k] > maxVal: 
        maxVal = dict_5char[k]
        maxKey = k
```

```python
print(f'max 5-long-sequence is {maxKey} with frequency equals {maxVal}')
```
