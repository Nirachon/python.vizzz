## Add column from another table ตัดตารางนึงมาแปะอีกตารางหนึ่ง

## .map()

### add business name to search_click

```python
search_click_df.head()
```

```python
search_click_df['business_name'] = search_click_df['business_id'].map(small_table.set_index('id')['name'])
```

ชื่อตารางตามด้วยชื่อคลัมใหม่ = ตารางหลัก+คีย์ตารางหลัก.ตารางตามด้วยคีญืารางรอง ต้องเชื่อมกันได้ 👆ชื่อของตารางรอง

###### เพิ่มคอลัมที่ชื่อว่า businessname ให้ค่าเท่ากับ
##### มีคีย์ชื่อบสนอด ไปแมบกับ...โดยคีย์ขอมันชื่อว่าไอดี ตารางหลักคือเปลี่ยนแปลง ตารางรองคือเอาค่ามา
หลักคีย์ซ้ำได้ รองห้ามซ้ำ

```python
search_click_df.head()
```

### combine two table รวมสองตารางเข้าด้วยกัน

```python
marged_table = search_click_df.merge(small_table, how = 'left',left_on = 'business_id' , right_on = 'id')
```

ตารางหลักจะเมิจกับตารางรอง มีพารามิเตอร์ชื่อฮาว ถามว่าตัวไหนเป็นตารางหลัก เราก็บอกว่าleft left_onคีย์ตารางด้านซ้ายคืออะไร
