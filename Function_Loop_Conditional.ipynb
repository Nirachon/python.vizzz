````python
list_grade =  [1,50,65,90,101,-5,49]
````

````python
for a in list_grade:
    if (a<=100) and (a>0):
        if(a>=80 ):
            print("Grade : A")
        elif (a>=75):
            print("Grade : B+")
        elif (a>=70):
            print("Grade : B")
        elif (a>=65):
            print("Grade : C+")
        elif (a>=60):
            print("Grade : C")    
        elif (a>=55):
            print("Grade : D+")
        elif (a>=50):
            print("Grade : D")
        elif (a>=49):
            print("Grade : F")
    else:
        print("Grade : No grade")
````
