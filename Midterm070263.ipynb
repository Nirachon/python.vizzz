## สร้าง function ชื่อ smartDict ซึ่งทำหน้าที่:
- function รับ input เป็น list 2 ตัวที่มีขนาดเท่ากัน โดย list ตัวแรกเป็น list ของตัวเลขlistที่สองเป็น list ของtext
- ถ้า ค่า list แรกเป็นเลขคี่  ให้หาค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 1 และ value จาก list ตัวที่ 2
- ถ้า ค่า list แรกเป็นเลขคู่  ให้หาค่าใน dict ที่เป็น output มี key เป็นค่าจาก list ตัวที่ 2 และ value จาก list ตัวที่ 1
- output ของ function คือ dict ที่สร้างขึ้น


เช่น 
- output1 = smartDict([1,2,3],['apple','banana','orange'])
- จะได้ output1 เป็น{1:'apple','banana':2,3:'orange'}

```python
def   smartDict(listnumber,listtext):  
        if listnumber==  :        
                retdict ={ }
                for index in range(len(listnumber)) : 
                    retdict[listnumber[index]] = listtext[index]
        else:
                retdict ={ }
                for index in range(len(listnumber)) :
                     retdict[listtext[index]] = listnumber[index]
        return retdict
                 
```

#### 2.1 กรดอะมิโนใด(A,B,G,C) มีปริมาณมากสุด และมีเท่าไหร่

```python
import pandas as pd
```

```python
DNAseq = pd.read_csv('C:\\Users\\Biw\\Documents\\viz_python\\python_dataviz\\Wuhan_nCoV_GNome.txt',header = None)
```

```python
DNAseq.head()
```

```python
def countChar(seq,character): #row1,ตัอักษรแต่ละตัว
    charcount = 0 
    for char in seq:
        if char == character: 
            charcount = charcount + 1 # เอาตัวที่นับเจอไปบวกกับตัวนับไปแล้ว
    return charcount
```

```python
ANum = 0 #หาความยาว
BNum = 0
GNum = 0
CNum = 0
for row in range (DNAseq.shape[0]):      #rang (len(DNAseq))
    ANum += countChar(DNAseq.iloc[row][0],'A')
    TNum += countChar(DNAseq.iloc[row][0],'T')
    GNum += countChar(DNAseq.iloc[row][0],'G')
    CNum += countChar(DNAseq.iloc[row][0],'C')

print(f'There are  {ANum} A, {TNum} B , {GNum} G , {CNum} C ')
```

### 2.2 sequence ที่มีขนาด 3 และ 4 ที่เกิดซ้ำมากที่สุด

### ซ้ำ3ตัว

```python
def populatate3charstring(seq):
    str_list = list()
    for i in range(len(seq)-2) : 
        str_list.append(seq[i:i+3]) 
        
    return str_list
```

```python
LIST_ALL_3CHAR = list() 
for row in range (DNAseq.shape[0]):
    LIST_ALL_3CHAR = LIST_ALL_3CHAR + populatate3charstring(DNAseq.iloc[row][0])
```

```python
len(LIST_ALL_3CHAR)
```

```python
unique_3char = (set(LIST_ALL_3CHAR))
```

```python
len(unique_3char)
```

```python
dict_3char = {}
for sub_str in unique_3char: 
    num1 = 0
    for s in LIST_ALL_3CHAR: 
        if sub_str == s: 
            num1 += 1
    dict_3char[sub_str] = num1
```

```python
dict_3char
```

```python
maxVal1 = -100 
maxKey1 = None 
for i in dict_3char.keys(): 
    if dict_3char[i] > maxVal1: 
        maxVal1 = dict_3char[i]
        maxKey1 = i
```

```python
maxKey1
```

```python
print(f'max 3-long-sequence is {maxKey1} with frequency equals {maxVal1}')
```
