# Radar chart

```python
from math import pi
```

```python
import pandas as pd
```

```python
data = [('c1', 2, 2,5,4,2),('c2', 5, 1,3,3,3),('c3', 4, 3, 3,2,5),('c4', 3, 5, 4,1,3)]
labels = ['client_id' , 'cheese','beer', 'chicken','bread', 'coffee']
df = pd.DataFrame.from_records(data, columns=labels)
```

```python
df
```

```python
categories = list(df)[1:]
```

```python
categories 
```

```python
values = df.mean().values.flatten().tolist()
values += values[:1] # repeat the first value to close the circular graph
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]
```

```python
values 
```

```python
angles
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8), #วาดวงกลมให้พร้อม
                       subplot_kw=dict(polar=True))
plt.xticks(angles[:-1], categories, color='grey', size=12) #เปลี่ยนหน่วย
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)
plt.ylim(0, 5)
ax.set_rlabel_position(30)
ax.plot(angles, values, linewidth=1, linestyle='solid')
ax.fill(angles, values, 'skyblue', alpha=0.4) ;
```

ใช้สับพอตสร้างเรด้า

```python
# preparre canvas
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8), #วาดวงกลมให้พร้อม
                       subplot_kw=dict(polar=True))
plt.xticks(angles[:-1], categories, color='grey', size=12) #เปลี่ยนหน่วย
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

#part2
val_c1 = df.loc[0].drop('client_id').values.flatten().tolist()
val_c1 += val_c1[:1]
ax.plot(angles, val_c1, linewidth=1,
        linestyle='solid', label='Client c1')
ax.fill(angles, val_c1, 'skyblue', alpha=0.4)
 
# part2
val_c2=df.loc[1].drop('client_id').values.flatten().tolist()
val_c2 += val_c2[:1]
ax.plot(angles, val_c2, linewidth=1,
        linestyle='solid', label='Client c2')
ax.fill(angles, val_c2, 'lightpink', alpha=0.4)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```

# สร้าง Word cloud ของคอลัมน์

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
text = " ".join(review for review in word.message)
print("There are {} words in the combination of all review.".format(len(text)))  # ดึงคอลัมน์ที่ต้องการ
wordcloud = WordCloud(font_path='THSarabun Bold.ttf',regexp = r"[ก-๙a-zA-Z']+" #เอาฟอนต์ตัวอักษรไปวางที่ไฟล์ที่เราเปิดjupyterก่อน
                    ,width=2000, height=1500, margin=0,
                      colormap='plasma',max_words=2000,min_font_size=4,background_color="white").generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```
