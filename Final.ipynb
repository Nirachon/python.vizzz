# สร้างคอลัมน์ใหม่คือเวลาและวันออกจากคอลัมน์Date

```python
data = pd.DataFrame(df)
```

```python
data[['Day','Time']] = df.Date.apply( 
   lambda x: pd.Series(str(x).split(" "))) 
   
print(data) 
```
