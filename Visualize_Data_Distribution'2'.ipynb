พอตBOX PLOT ของตัว setosa จะได้ทั้งหมดกี่อัน

```python
setosa_np =irisdata.iloc[:50,:-1].to_numpy()
vericolor_np =irisdata.iloc[50:100,:-1].to_numpy()
virginica_np =irisdata.iloc[100:150,:-1].to_numpy()
```

```python
plt.rcParams['figure.figsize']=[12,12]
plt.subplot(1,4,1)
plt.boxplot(setosa_np[:,0])
plt.subplot(1,4,2)
plt.boxplot(setosa_np[:,1])
plt.subplot(1,4,3)
plt.boxplot(setosa_np[:,2])
plt.subplot(1,4,4)
plt.boxplot(setosa_np[:,3])
plt.show()
```

### use boxplot to compare features between 3 classes เปรียบเทียบฟีเจอร์แต่ละคลาส

```python
plt.rcParams['figure.figsize']=[16,16]
col1se=setosa_np[:,0]
col1ve=vericolor_np[:,0]
col1vi=virginica_np[:,0]
plt.boxplot((col1se, col1ve,col1vi),labels=['setosa','vericolor','virginica'])
plt.title('Feature 1 comparison');
```

```python
plt.rcParams['figure.figsize']=[16,16]
col2se=setosa_np[:,1]
col2ve=vericolor_np[:,1]
col2vi=virginica_np[:,1]
plt.boxplot((col2se, col2ve,col2vi),labels=['setosa','vericolor','virginica'])
plt.title('Feature 2 comparison');
```

```python
plt.rcParams['figure.figsize']=[16,16]
col3se=setosa_np[:,2]
col3ve=vericolor_np[:,2]
col3vi=virginica_np[:,2]
plt.boxplot((col3se, col3ve,col3vi),labels=['setosa','vericolor','virginica'])
plt.title('Feature 1 comparison');
```

```python
plt.rcParams['figure.figsize']=[16,16]
col4se=setosa_np[:,3]
col4ve=vericolor_np[:,3]
col4vi=virginica_np[:,3]
plt.boxplot((col4se, col4ve,col4vi),labels=['setosa','vericolor','virginica'])
plt.title('Feature 1 comparison');
```
